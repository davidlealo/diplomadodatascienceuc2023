# Estudiantes desertores por año
Prueba de función de estudiantes desertores por año

```python
# Función de estudiates desertores --- PROBLEMA no ELIMINA SOLAMENTE 4 medio, sino también 4 básico y otros grados 

def identificar_desertores(archivo_anterior, archivo_actual):
    # Cargar los archivos CSV en DataFrames de Pandas
    df_anterior = archivo_anterior
    df_actual = archivo_actual

    # Filtrar los estudiantes que tienen un código de grado inferior a 12
    desertores = df_anterior[df_anterior['COD_GRADO'] != 4]

    # Verificar si los estudiantes están presentes en el archivo actual
    desertores = desertores[~desertores['MRUN'].isin(df_actual['MRUN'])]

    # Devolver el resultado como DataFrame de Pandas
    return desertores

```

```python
# Segunda versión de función estudiantes desertores


def identificar_desertores2(df_anterior, df_posterior):
    # Filtrar los estudiantes que tienen un código de grado inferior a 4 medio
    desertores = df_anterior[~((df_anterior['COD_GRADO'] == 4) & (df_anterior['COD_ENSE2'].isin([5, 7]))) & ~(df_anterior['COD_GRADO'] > 9)]

    # Verificar si los estudiantes están presentes en el archivo actual
    desertores = desertores[~desertores['MRUN'].isin(df_posterior['MRUN'].unique())]

    # Devolver el resultado como DataFrame de Pandas
    return desertores
```

```python
# Tercera versión de función estudiantes desertores


def identificar_desertores3(df_anterior, df_posterior):
    # Filtrar los estudiantes que no son prebásica o de educación para adultos
    desertores = df_anterior[(df_anterior['COD_ENSE2'].isin([1, 6, 8]))]

    # Filtrar los estudiantes que tienen un código de grado inferior a 4 medio
    desertores = df_anterior[~((df_anterior['COD_GRADO'] == 4) & (df_anterior['COD_ENSE2'].isin([5, 7]))) & ~(df_anterior['COD_GRADO'] > 9)]

    # Verificar si los estudiantes están presentes en el archivo actual
    desertores = desertores[~desertores['MRUN'].isin(df_posterior['MRUN'].unique())]

    # Devolver el resultado como DataFrame de Pandas
    return desertores
```

```python

m2018 = pd.read_csv('data/matriculas/2018/m2018.csv', sep=';')
m2019 = pd.read_csv('data/matriculas/2019/m2019.csv', sep=';')
total2019 = (m2019['COD_ENSE2'].isin([2, 3, 4, 5, 6, 7, 8]).sum())

# Identificar desertores
desertores_2018_2019 = identificar_desertores3(m2018, m2019)
print("Total de desertores:",len(desertores_2018_2019))
print("Desertores por género del estudiante")
print("Sin información:",((desertores_2018_2019['GEN_ALU'] == 0).sum()))
print("Hombre:",((desertores_2018_2019['GEN_ALU'] == 1).sum()))
print("Mujer:",((desertores_2018_2019['GEN_ALU'] == 2).sum()))
print("Desertores por tipo de dependencia")
print("Municipal:",((desertores_2018_2019['COD_DEPE2'] == 1).sum()))
print("Particular subvencionado:",((desertores_2018_2019['COD_DEPE2'] == 2).sum()))
print("Particular pagado:",((desertores_2018_2019['COD_DEPE2'] == 3).sum()))
print("Corporación de administración delegada:",((desertores_2018_2019['COD_DEPE2'] == 4).sum()))
print("Servicio local de educación:",((desertores_2018_2019['COD_DEPE2'] == 5).sum()))
print("Desertores por tipo de enseñanza")
print("Educación básica:",((desertores_2018_2019['COD_ENSE2'] == 1).sum()))
print("Educación media humanístico-científica:",((desertores_2018_2019['COD_ENSE2'] == 2).sum()))
print("Educación media técnico profesional:",((desertores_2018_2019['COD_ENSE2'] == 3).sum()))
print("Educación para adultos:",((desertores_2018_2019['COD_ENSE2'] == 6).sum()))
print("Educación prebásica:",((desertores_2018_2019['COD_ENSE2'] == 8).sum()))
```
